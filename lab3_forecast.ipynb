```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
dataset = pd.read_csv("C:/Users/hevas/Downloads/lab3_forecast/train.csv")
dataset.head()
```

```python
dataset['aniomes']   = dataset['fecha'].dt.strftime('%Y%m')
dataset['anio']      = dataset['fecha'].dt.strftime('%Y')
dataset['codfecha']   = dataset['mes'].dt.strftime('%Y%m%d')
dataset['trimestre'] = dataset['fecha'].dt.quarter
dataset
```

```python
dataset.set_index("date", inplace=True)
```

```python
plt.figure(figsize=(12,6))
plt.plot(dataset.index,dataset['sales'],'-o')

plt.xlabel('fecha',fontsize=14)
plt.ylabel('Cantidad',fontsize=14)
plt.title('sales por Mes', fontsize=18)
plt.show()
```

```python
ax = sns.distplot(dataset['sales'], fit=norm, kde=False)
```

```python
dataset['store'].value_counts().sort_values(ascending=False).plot.bar()
plt.ylabel("Total store")
plt.xlabel("store")
plt.show()
```

```python
dataset['item'].value_counts().sort_values(ascending=False).plot.bar()
plt.ylabel("Total item")
plt.xlabel("item")
plt.show()
```

```python
sns.factorplot('item', data=dataset, hue='Class',kind="sum")

dataset[['Class','item']].groupby(['Class'], as_index=False).agg(['count','sum','mean'])
```
